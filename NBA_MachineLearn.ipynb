**Setting predictors**

```python
predictors = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year',
       'W', 'L', 'W/L%', 'GB', 'PS/G',
       'PA/G', 'SRS'] ## using 'Share' will give the algorithm direct values to predict what we want, Share gives percentage of MVP won which is what we are predicting so remove any columns which
                      # affect Share like 'Pts Won', 'Pts Max'.
```

**Train data on anything before 2021, and test data on 2021**

```python
train = stats[stats["Year"] < 2021]
```

```python
test = stats[stats["Year"] == 2021]
# Data is temporal so data should be used to predict what is happening at the end of a season. so in this case using it to predict on 2021
# testing should be the same of how we want to use the algorithm in the real world
# When working with "time series" data, dont use data from after your test set
# If you do it will "overfit", so it will do well in training but fail in testing
# So having a test in your dataset is very useful
```

**Error Metric for correct Top 5 placements**

```python
# Give a postive score if the predictor place the player in the top 5 in top 5, penalise the predictor for how far it took to place that player
def find_ap(combination):
  # Give a point every time a player is in the top 5, if they are not it gets penalised for not being in the top 5
  actual = combination.sort_values("Share", ascending=False).head(5) #Get the top 5 of actual winners
  predicted = combination.sort_values("predictions", ascending=False)
  ps = []
  found = 0
  seen = 1
  for index, row in predicted.iterrows(): #go through each row in the predicted table
    if row["Player"] in actual["Player"].values: #if the player exists in the top 5, give a point on found
      found += 1
      ps.append(found/seen) #appends the precentage for how long it took to find the player in top 5
    seen += 1 #If the player was not in the top 5, penalise
  return sum(ps) / len(ps) # Gives an error metric
```

```python
find_ap(combination) # Closer to 1 the better
```

**Backtesting for most years**

```python
years = list(range(1991, 2022))
```

```python
aps = []
all_predictions = []
for year in years[5:]: # need some data to make predictions with in order to make predictions
  train = stats[stats["Year"] < year] #Iterate through each year, use previous years as training data
  test = stats[stats["Year"] == year] #Then us the current year for testing

  reg.fit(train[predictors], train["Share"])
  predictions = reg.predict(test[predictors])
  predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
  combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)

  all_predictions.append(combination)
  aps.append(find_ap(combination))
```

**Mean average precision**

```python
sum(aps) / len(aps)
```

**Diagnosis of Algorithm**

```python
def add_ranks(combination): # Having good diagnostic info makes diagnosing and improving the algorithm easier
  combination = combination.sort_values("Share", ascending=False)
  combination["Rk"] = list(range(1, combination.shape[0]+1))

  combination = combination.sort_values("predictions", ascending=False)
  combination["Predicted_Rk"] = list(range(1, combination.shape[0]+1))

  combination["Difference"] = combination["Rk"] - combination["Predicted_Rk"]
  return combination
```

```python
ranking = add_ranks(all_predictions[1])
ranking[ranking["Rk"] <= 5].sort_values("Difference", ascending=False)
```

**Turning backtest into function**

```python
def backtest(stats, model, year, predictors): #Making a function so it is easier to call the backtesting we did earlier
  aps = []
  all_predictions = []
  for year in years[5:]:
    train = stats[stats["Year"] < year]
    test = stats[stats["Year"] == year]

    model.fit(train[predictors], train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    combination = add_ranks(combination)

    all_predictions.append(combination)
    aps.append(find_ap(combination))
  return sum(aps)/len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, aps, all_predictions = backtest(stats, reg, years, predictors)
```
